### Insight 1 Explanation

Ranking Deaths and Cases by State

We wanted to compare numbers in regards to COVID-19 cases and deaths based on the states, to see where states ranked among each other in proportions of cases and deaths, respectively. Locale of the COVID hotspots is a big concern for most people, so to see the face value of such distribution would be eye-catching.

From the analysis, we found the results reflected in the table. Any "New" category is defined as the increment from one day/data entry to the next. So, "Highest New Cases" represents the highest found one-day increases in cases.

 We can observe that California and Texas tend to linger in the apex of the "Highest Categories", while states like Hawaii, Vermont, Wyoming, Maine, and Arkansas reside in the tail end. This should be taken with consideration of the respective populations of the states, which explain their trends to a degree.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
def insight2(filename):
  data = pd.read_csv(filename, index_col = 0)
  data["Date"] = pd.to_datetime(data["Date"])
  df1 = data.groupby(pd.Grouper(key = "Date", freq = '1M'))["Adults in Hosp"].mean()
  df1.index = df1.index.strftime("%B %Y")

  df2 = data.groupby(pd.Grouper(key = "Date", freq = '1M'))["Children in Hosp"].mean()
  df2.index = df2.index.strftime("%B %Y")

  master = pd.DataFrame([df1, df2])
  master.dropna(axis=1, inplace = True)
  master = master.transpose()
  master["Date"] = master.index
  master = master.iloc[3:]


  master.drop("Date", axis = 1, inplace = True)
  return np.round(master, 2)



############ Function Call ############
insight2("Cleaned_Hosp.csv")
```

### Insight 2 Explanation

Average COVID In-Patients at a Time

Hospital capacity has been one of the most serious issues in the wake of the pandemic, so an interesting statistic would relate to the number of COVID patients hospitalized, or in-patients, at a time.

By taking the mean of the values for the number of adults/children hospitalized with COVID on a monthly basis across all data entries from all states, the table produces yields the expected value of the number of COVID in-patients hospitalized at any given time, within the month, in any state, categorized by adults and children.

The values of these measurements, for both children and adults, are exceedingly large in the months of July and August of 2020, stay relatively high, increase again at February 2021, and then dip drammatically onwards.

It is also noticeable that the patterns in the fluctuations of the numbers in the "Children" category reflect the patterns of the "Adults" category well, since peaks and troughs are roughly simultaneous.

### Insight 3 Explanation

F:V Ratio

Since much research has been done for the purpose of the second administration of certain COVID vaccines, the statistics regarding full vaccinations among people would be an interesting angle to view.

The table shows the monthly national ratio of People Fully Vaccinated to People Vaccinated. We can see the that the ratio increases dramatically in the first few months, which is to be expected as this is when the vaccine is in the beginning of its public exposure.

The decrease in the increment of this ratio shows that a smaller percentage of those who are vaccinated are not fully vaccinated, either because many have already gotten the first administration, or less people are left to get a first administration.

However, the latter situation may not be the case, since the ratio decreases from July to August, likely indicating a rise in people getting their first administrations. As a result, it would be interesting to study cohort factors that may have influenced this change. One of those factors may be the extension of vaccine administration to younger ages in the tail end of August.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
def summary1():
  d1 = pd.read_csv("Cleaned_Deaths_and_Cases.csv")
  d2 = pd.read_csv("Cleaned_Hosp.csv")
  d3 = pd.read_csv("Cleaned_Vaccinations.csv")
  d4 = pd.read_csv("Cleaned_Recoveries.csv")

  d1["Date"] = pd.to_datetime(d1["Date"])
  d2["Date"] = pd.to_datetime(d2["Date"])
  d2["Total in Hosp"] = d2["Adults in Hosp"] + d2["Children in Hosp"]
  d3["Date"] = pd.to_datetime(d3["Date"])
  d4["Date"] = pd.to_datetime(d4["Date"])
  d3 = d3[d3["Daily Vaccinations"] != -1.0]

  e3 = d3.groupby(pd.Grouper(key = "Date", freq = '1D'))["People Vaccinated"].sum()
 # e3.index = e3.index.strftime("%b %d %Y")
  master3 = pd.DataFrame([e3]).transpose()

  e1 = d1.groupby(pd.Grouper(key = "Date", freq = '1D'))["Total Deaths"].sum()
 # e1.index = e1.index.strftime("%b %d %Y")
  master1a = pd.DataFrame([e1]).transpose()

  f1 = d1.groupby(pd.Grouper(key = "Date", freq = '1D'))["Total Cases"].sum()
 # f1.index = f1.index.strftime("%b %d %Y")
  master1b = pd.DataFrame([f1]).transpose()

  e4 = d4.groupby(pd.Grouper(key = "Date", freq = '1D'))["Total Recoveries"].sum()
  #e4.index = e4.index.strftime("%b %d %Y")
  master4 = pd.DataFrame([e4]).transpose()

  e2 = d2.groupby(pd.Grouper(key="Date", freq = '1D'))["Total in Hosp"].sum()
  master2 = pd.DataFrame([e2]).transpose()
  master2 = master2[master2["Total in Hosp"] != 0.0]

  elem = master3.merge(master4, how ="right", on = "Date")

  flem = elem.merge(master1a, how = "left", on ="Date")

  ylem = flem.merge(master1b, how = "left", on = "Date")

  glem = ylem.merge(master2, how ="left", on = "Date")


  glem = glem[glem["People Vaccinated"] != 0.0]

  glem["Date"] = glem.index

  pd.set_option("display.max_rows", None)


  glem = glem[glem["Date"].isin(["Feb 21 2021", "Feb 23 2021", "Feb 24 2021", "Mar 23 2021", "Apr 12 2021", "Apr 13 2021", "Jun 11 2021", "Jun 12 2021",
                                 "Jun 13 2021", "Jun 14 2021", "Jun 15 2021", "Jun 16 2021", "Jun 17 2021", "Jun 18 2021", "Jun 19 2021",
                                 "Jun 20 2021", "Jun 21 2021", "Jun 22 2021", "Jun 23 2021", "Jun 24 2021", "Jul 16 2021", "Jul 17 2021", "Jul 18 2021", "Jul 19 2021",
                                 "Aug 09 2021", "Aug 10 2021", "Aug 14 2021", "Aug 15 2021", "Aug 16 2021", "Aug 23 2021", "Aug 24 2021", "Aug 25 2021", "Aug 26 2021",
                                 "Aug 27 2021", "Aug 28 2021", "Aug 29 2021"]) == False]

  glem.fillna("N/A", axis = 1, inplace = True)

  pd.set_option('display.float_format', lambda x: '%.0f' % x)

  glem.drop("Date", axis = 1, inplace = True)

  glem.to_csv("Combined_Covid_Summaries.csv", encoding = "utf-8")



############ Function Call ############
summary1()
```
